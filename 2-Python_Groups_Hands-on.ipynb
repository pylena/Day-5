#### Q8: Create a function to bulk update genres of books using list comprehension. and test it.

```python
# write your code here ^_^

def updateB(oldGenra:str,newGenra:str, bookDict:dict):
    for book in bookDict.values():
       book['genres'] = [
           newGenra if genra == oldGenra else genra for genra in book['genres']]

updateB('Fiction','NewF', library)  
print(library)    
    
```
